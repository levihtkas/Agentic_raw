```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
open_api_key = os.getenv("OPENAI_API_KEY")

anthropic_api_key = os.getenv("ANTHROPIC_API_KEY")

from openai import OpenAI

client_openai = OpenAI(api_key=open_api_key)
client_anthropic = OpenAI(api_key=anthropic_api_key,base_url="https://api.anthropic.com/v1")


```

```python
# 
def messages_for(user_message):
    
    return [        {"role":"system",
        "content": f"You are a helpful bussiness analyst."

    },{
        "role": "user",
        "content": f"{user_message}"}]
```

```python
#exercise 

response = client_openai.chat.completions.create(model="gpt-5-mini",messages=messages_for("Pick up a bussiness area might be worth for agentic ai opportunity"))
reply = response.choices[0].message.content
print(reply)
response_2 = client_openai.chat.completions.create(model="gpt-5-mini",messages=messages_for(reply+" Now, for this problem a pain-point in that industry - something challenging that might be ripe for an Agentic solution "))
reply_2 = response_2.choices[0].message.content
print(reply_2)
response_3 = client_openai.chat.completions.create(model="gpt-5-mini",messages=messages_for(reply+reply_2+"propose the Agentic AI solution"))
reply_3 = response_3.choices[0].message.content    
print(reply_3)
```
