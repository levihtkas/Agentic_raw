```python
request = "Please come up with a challenging, nuanced question that I can ask a number of LLMs to evaluate their intelligence. "
request += "Answer only with the question, no explanation."
messages = [{"role": "user", "content": request}]
```

```python
# 5 workflow design patterns
#1. Prompt Chaining
#2. Routing
#3. Parallelism
#4. Orchestration
#5. Evaluator-Optimizer 

#1. Prompt Chaining
#  What this would mean is that there will be gates in between which will have that logic before it passes into the next LLM. LLMs will be chained together by the different system prompts, and they will be given the task to complete.


def messages_for(prompt):
    return [{"role": "user", "content": prompt}]


def prompt_chaining(msg:str)->str:
    messages = messages_for(msg)
    response_1 = client_openai.chat.completions.create(model="gpt-5-mini",messages=messages)
    reply_1 = response_1.choices[0].message.content
    
    print("***OPENAI DONE***")
    reply_1= f"\n\n here is the reply from another LLM. Your job is to refine it and answer the questions from the previous LLM, thinking about what needs to be improved and giving a set of questions to the next LLM. The final LLM will be able to make a correct decision based upon that. \n response:{reply_1}" 
    print(reply_1)
    response_2 = client_openai.chat.completions.create(model="gpt-5-mini",messages=messages_for(reply_1))
    reply_2 = response_2.choices[0].message.content
    reply_2 = f"here is the response of the previous LLM, and your job is to rank these projects. First, answer these questions and rank these projects according to which you think will be suitable for the developer{reply_2}"
    print("***OPENAI 2 DONE***")
    response_3 = client_anthropic.chat.completions.create(model="claude-sonnet-4-5-20250929",messages=messages_for(reply_1+reply_2))
    reply_3 = response_3.choices[0].message.content
    return reply_3

message = "We do come up with a few projects which could take any engineer to a level that will push a computer science engineer to the next level. It must be in the fields of HNTK and AI, and it should have a slight nuance in system design so that the developer gets a feel for it. What you should be doing is you should be coming up with a series of questions that will make the other LLM think about how to decide which project to take."
prompt_chaining(message)
```
